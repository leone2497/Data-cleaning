```python
import pandas as pd
import numpy as np
from google.colab import drive
```

```python
Mattia_calendar= pd.read_excel('/content/drive/MyDrive/Dati_lavoro/Mattia calendar.xlsx')
Luigi_calendar= pd.read_excel('/content/drive/MyDrive/Dati_lavoro/Luigi calendar.xlsx')
```

```python
def calendar_converter_format(start,end):
   end_up=end.str.split().str[1:3]
   start_up=start.str.split().str[1:3]
   return start_up, end_up



```

```python
calendar_converter_format(Mattia_calendar['Start'],Mattia_calendar['End'])
Mattia_calendar['Start'],Mattia_calendar['End']= calendar_converter_format(Mattia_calendar['Start'],Mattia_calendar['End'])
```

```python
calendar_converter_format(Luigi_calendar['Start'],Luigi_calendar['End'])
Luigi_calendar['Start'],Mattia_calendar['End']= calendar_converter_format(Luigi_calendar['Start'],Luigi_calendar['End'])
```

```python
start_date = '2022-01-01'
end_date = '2024-12-31'
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
hours = pd.timedelta_range(start='9 hours', end='18 hours', freq='H')
date_hour_combinations = pd.DataFrame([(date + hour) for date in date_range for hour in hours], columns=['Date'])
```

```python
def convert_to_datetime(date_time_list):
    if isinstance(date_time_list, list):
        return pd.to_datetime(' '.join(date_time_list), format='%d/%m/%Y %H:%M')
    else:
        return pd.NaT
```

```python
for columns in Mattia_calendar.columns:
    Mattia_calendar[columns]=Mattia_calendar[columns].apply(convert_to_datetime)
```

```python
for columns in Luigi_calendar.columns:
    Luigi_calendar[columns]=Luigi_calendar[columns].apply(convert_to_datetime)
```

```python
Mattia_calendar['Date']= Mattia_calendar['Start']
Luigi_calendar['Date']= Luigi_calendar['Start']
```

```python
Mattia_calendar
```

```python
Calendario = pd.merge(date_hour_combinations , Mattia_calendar, how='outer', on='Date')
```

```python
Calendario = pd.merge(Calendario , Luigi_calendar, how='outer', on='Date')
```

```python
Calendario=Calendario.fillna('Free')
Calendario=Calendario.set_index('Date')
```

```python
Free= Calendario.index[(Calendario== 'Free').all(axis=1)]
```
